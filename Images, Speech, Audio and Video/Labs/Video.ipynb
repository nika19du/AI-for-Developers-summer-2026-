```python
from lumaai import LumaAI
from google.colab import userdata

api_key = userdata.get("LUMAAI_API_KEY")

client = LumaAI(auth_token = api_key)
```

```python
import time

def wait_until_completed(generation, max_attempts=100, delay_seconds=2):
    for _ in range(max_attempts):
        generation = client.generations.get(id=generation.id)

        print("State:", generation.state)

        if generation.state == "completed":
            return generation

        if generation.state == "failed":
            raise RuntimeError(
                f"Generation failed: {generation.failure_reason}"
            )

        time.sleep(delay_seconds)

    raise RuntimeError(
        f"Video generation couldn't finish within "
        f"{max_attempts * delay_seconds} seconds."
    )
```

```python
from IPython.display import Image, Markdown, Video

def display_generation(generation):
  print(f"Generation id: {generation.id}")

  if generation.assets.image is not None:
    display(Markdown("### Thumbnail"))
    display(Image(url=generation.assets.image))

  display(Markdown("### Video"))
  display(Video(generation.assets.video))
```

## Text to Video

```python
funny_cat_prompt = "Generate a short and funny video of a cat that is playing with its tail, while laying on a mat."
```

```python
funny_cat_generation = client.generations.create(
    prompt = funny_cat_prompt,
    model = "ray-2",
    aspect_ratio = "16:9",
    duration = "5s",
    resolution= "540p"
)
```

```python
funny_cat_generation = wait_until_completed(funny_cat_generation)
```

```python
display_generation(funny_cat_generation)
```

## Text - Image to Video

```python
walking_cat_prompt = "Extend this scene naturally. A realistic orange tabby cat calmly walks across the kitchen counter from left to right with smooth, natural movements, occasionally looking around and gently swishing its tail. Keep the original kitchen, lighting, composition, and camera angle unchanged. Animate only the cat while preserving all other objects. The motion should be photorealistic, cinematic, stable, and physically accurate."
kitchen_scene_url = "https://images.pexels.com/photos/2089698/pexels-photo-2089698.jpeg"
```

```python
walking_cat_generation = client.generations.create(
    prompt = walking_cat_prompt,
    model = "ray-flash-2",
    aspect_ratio = "16:9",
    duration = "5s",
    resolution = "540p",
    keyframes = {
        "frame0":{
            "type": "image",
            "url": kitchen_scene_url
        }
    }
)
```

```python
walking_cat_generation=wait_until_completed(walking_cat_generation)
```

```python
walking_cat_generation
```

```python
display_generation(walking_cat_generation)
```
